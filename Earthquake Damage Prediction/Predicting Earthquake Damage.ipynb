### Importing Library

```python
import pandas as pd
import numpy as np

train_values = pd.read_csv('train_values_DD.csv')
train_labels = pd.read_csv('train_labels_DD.csv')
test_values = pd.read_csv('test_values_DD.csv')

train_values.head()
```

```python
# drop feature
train_values.drop(['building_id'], axis = 1, inplace = True)
test_values.drop(['building_id'], axis = 1, inplace = True)
```

```python
train_labels.drop(['building_id'], axis = 1, inplace = True)
```

```python
# converting to categorical variable
train_values['land_surface_condition'] = pd.Categorical(train_values['land_surface_condition'])
train_values['foundation_type'] = pd.Categorical(train_values['foundation_type'])
train_values['roof_type'] = pd.Categorical(train_values['roof_type'])
train_values['ground_floor_type'] = pd.Categorical(train_values['ground_floor_type'])
train_values['other_floor_type'] = pd.Categorical(train_values['other_floor_type'])
train_values['position'] = pd.Categorical(train_values['position'])
train_values['plan_configuration'] = pd.Categorical(train_values['plan_configuration'])
train_values['legal_ownership_status'] = pd.Categorical(train_values['legal_ownership_status'])
```

```python
# converting to categorical variable
test_values['land_surface_condition'] = pd.Categorical(test_values['land_surface_condition'])
test_values['foundation_type'] = pd.Categorical(test_values['foundation_type'])
test_values['roof_type'] = pd.Categorical(test_values['roof_type'])
test_values['ground_floor_type'] = pd.Categorical(test_values['ground_floor_type'])
test_values['other_floor_type'] = pd.Categorical(test_values['other_floor_type'])
test_values['position'] = pd.Categorical(test_values['position'])
test_values['plan_configuration'] = pd.Categorical(test_values['plan_configuration'])
test_values['legal_ownership_status'] = pd.Categorical(test_values['legal_ownership_status'])
```

```python
train_values.dtypes
```

```python
# convert all the features to numeric using OneHotEncoding
train_values = pd.get_dummies(train_values, drop_first = True)
```

```python
# convert all the features to numeric using OneHotEncoding
test_values = pd.get_dummies(test_values, drop_first = True)
```

```python
train_values.head()
```

```python
from sklearn.model_selection import train_test_split

X = train_values
y = train_labels

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=300, max_depth=200, random_state=0)
rf.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

scores = cross_val_score(rf, X_train, y_train, cv=10)
print('CV accuracy: %.3f +/- %.3f' % (np.mean(scores), np.std(scores)))
```

```python
from sklearn.metrics import mean_absolute_error, accuracy_score, classification_report

y_test_pred = rf.predict(X_test)

print(classification_report(y_test, y_test_pred))
print('Accuracy:', accuracy_score(y_test, y_test_pred))
print('MAE:', mean_absolute_error(y_test, y_test_pred))
```
